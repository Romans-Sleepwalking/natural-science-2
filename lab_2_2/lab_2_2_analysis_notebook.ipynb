# Summary

* **There is not field inside both rings.** That proves that both rings are metal conductors and not plastic isolators. The electric field inside the rings is zero because the electrons accumulate inside the ring moving around; therefore, cancelling the field inside the ring.


```python
%load_ext autoreload
%autoreload 2

from measurements import LINES, POINTS, SQUARE, CIRCLES

import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

### 1. Define shapes
1. n - equipotential lines for two given electrodes.
2. m - equipotential lines when a conductor and/or an isolator are placed in the field.


```python
fig, ax = plt.subplots(figsize=(8, 10))

for line in LINES:
    plt.plot(line[:, 0], line[:, 1], marker='o', color='C0', linestyle='--', label="Equipotential line")

points_x, points_y = zip(*POINTS)
plt.scatter(points_x, points_y, color='C0', marker='o', label="Equipotential line")

# Draw square object above center
square_coords = (SQUARE[0], SQUARE[1])
square = patches.Rectangle(square_coords, SQUARE[2], SQUARE[2], fill=True, color='C2', linewidth=3, label="Electrode")
ax.add_patch(square)

for circle in CIRCLES:
    circle_coords = (circle[0], circle[1])
    plt_circle = patches.Circle(circle_coords, circle[2], fill=False, color='C3', linewidth=3, label="Either conductor or isolator")
    ax.add_patch(plt_circle)

ax.set_xlim(-4, 9)
ax.set_ylim(-7, 10)
ax.set_title("Reconstruction of the electric field and equipotential lines from protocol")
ax.grid(True)
handles, labels = ax.get_legend_handles_labels()
unique = dict(zip(labels, handles))
ax.legend(unique.values(), unique.keys())
plt.show()
```

The dashed blue curves show equipotential lines.
These lines are roughly vertical and curved, spacing varies depending on field strength.
The lines are denser near the positive electrode: green square.
The lines are more spread out on the opposite side, indicating a gradient in potential.

Two red circles are most likely conductors, since they disturb the field.
Red conductors distort equipotential lines by repelling or reshaping them.
Equipotential lines are around the circles.
Both rings are metallic and most metals are conductive.
Both rings are not isolators because the lines would pass through them and do not cause any distortion.

There is no electric field inside conducting rings.
If they were insulators, there could still be a field inside.
